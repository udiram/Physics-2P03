```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('accel_data.csv')
df = df[0:600] # truncate to remove noise
df.head()
```

```python
plt.plot(df['Time (s)'], df['Linear Acceleration z (m/s^2)'])
plt.xlabel('Time (s)')
plt.ylabel('Z-Axis Acceleration (m/s^2)')
plt.legend(['Z-Axis Acceleration'])
plt.show()
```

```python
#calculate velocity for each point using the z axis acceleration using a cumulative sum across the z axis acceleration
df['Velocity (m/s)'] = df['Linear Acceleration z (m/s^2)'].cumsum()
```

```python
plt.plot(df['Time (s)'], df['Velocity (m/s)'])
```

```python
#calculate velocity for each point using the z axis acceleration using a for loop
df['Velocity (m/s)'] = 0
for i in range(1, len(df)):
    df['Velocity (m/s)'][i] = df['Velocity (m/s)'][i-1] + df['Linear Acceleration z (m/s^2)'][i]
```

```python
plt.plot(df['Time (s)'] , df['Velocity (m/s)'])
```
