```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('accel_data.csv')
df_up = df[0:250] # truncate to remove noise
df_down = df[250:600]
df_up = df_up.reset_index(drop=True)
df_down = df_down.reset_index(drop=True)
```

```python
df_up.head()
```

```python
df_down.head()
```

```python
plt.plot(df['Time (s)'], df['Linear Acceleration z (m/s^2)'])
plt.xlabel('Time (s)')
plt.ylabel('Z-Axis Acceleration (m/s^2)')
plt.legend(['Z-Axis Acceleration'])
plt.show()
```

```python
plt.plot(df_up['Time (s)'], df_up['Linear Acceleration z (m/s^2)'])
plt.xlabel('Time (s)')
plt.ylabel('Z-Axis Acceleration (m/s^2)')
plt.legend(['Z-Axis Acceleration'])
plt.show()
```

```python
plt.plot(df_down['Time (s)'], df_down['Linear Acceleration z (m/s^2)'])
plt.xlabel('Time (s)')
plt.ylabel('Z-Axis Acceleration (m/s^2)')
plt.legend(['Z-Axis Acceleration'])
plt.show()
```

```python
#calculate velocity for each point using the z axis acceleration using a cumulative sum across the z axis acceleration
df_up['Velocity (m/s)'] = df_up['Linear Acceleration z (m/s^2)'].cumsum()
df_down['Velocity (m/s)'] = df_down['Linear Acceleration z (m/s^2)'].cumsum()
df['Velocity (m/s)'] = df['Linear Acceleration z (m/s^2)'].cumsum()
```

```python
plt.plot(df['Time (s)'], df['Velocity (m/s)'])
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend(['Velocity'])
```

```python
plt.plot(df_up['Time (s)'], df_up['Velocity (m/s)'])
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend(['Velocity'])
```

```python
plt.plot(df_down['Time (s)'], df_down['Velocity (m/s)'])
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend(['Velocity'])
```

```python
#get integral of velocity to get displacement
df_up['Displacement (m)'] = df_up['Velocity (m/s)'].cumsum()
df_down['Displacement (m)'] = df_down['Velocity (m/s)'].cumsum()
df['Displacement (m)'] = df['Velocity (m/s)'].cumsum()
```

```python
plt.plot(df['Time (s)'], df['Displacement (m)'])
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.legend(['Displacement'])
```

```python
plt.plot(df_up['Time (s)'], df_up['Displacement (m)'])
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.legend(['Displacement'])
```

```python
plt.plot(df_down['Time (s)'], df_down['Displacement (m)'])
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.legend(['Displacement'])
```

```python
# calculate the average velocity for the up and down motion
avg_velocity_up = df_up['Velocity (m/s)'].mean()
avg_velocity_down = df_down['Velocity (m/s)'].mean()
print("average velocity up: " + str(avg_velocity_up) + " m/s")
print("average velocity down: " + str(avg_velocity_down) + " m/s")
```

```python
# calculate the average displacement for the up and down motion
avg_displacement_up = df_up['Displacement (m)'].mean()
avg_displacement_down = df_down['Displacement (m)'].mean()
print("average displacement up: " + str(avg_displacement_up) + " m")
print("average displacement down: " + str(avg_displacement_down) + " m")
```

```python
# calculate the average acceleration for the up and down motion
avg_acceleration_up = df_up['Linear Acceleration z (m/s^2)'].mean()
avg_acceleration_down = df_down['Linear Acceleration z (m/s^2)'].mean()
print("average acceleration up: " + str(avg_acceleration_up) + " m/s^2")
print("average acceleration down: " + str(avg_acceleration_down) + " m/s^2")
```
