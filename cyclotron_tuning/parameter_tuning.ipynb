```python
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
source = df['source_current (A)']
foil = df['foil (uA)']
plt.plot(source, foil)
plt.xlabel('Source Current (A)')
plt.ylabel('Foil Current (uA)')
plt.title('Foil Current vs Source Current')
plt.savefig('foil.png')
print("Maximum of foil current is: ", foil.max(), "uA")
print("Maximum foil current occurs at: ", source[foil.idxmax()], "A")
```

```python
target = df['target (uA)']
plt.plot(source, target)
plt.xlabel('Source Current (A)')
plt.ylabel('Target Current (uA)')
plt.title('Target Current vs Source Current')
plt.savefig('target.png')
print("Maximum of target current is: ", target.max(), "uA")
print("Maximum target current occurs at: ", source[target.idxmax()], "A")
```

```python
column_1 = df['column 1 (uA)']
plt.plot(source, column_1)
plt.xlabel('Source Current (A)')
plt.ylabel('Column 1 Current (uA)')
plt.title('Column 1 Current vs Source Current')
plt.savefig('column_1.png')
print("Maximum of column 1 current is: ", column_1.max(), "uA")
print("Maximum column 1 current occurs at: ", source[column_1.idxmax()], "A")
```

```python
column_2 = df['column 2 (uA)']
plt.plot(source, column_2)
plt.xlabel('Source Current (A)')
plt.ylabel('Column 2 Current (uA)')
plt.title('Column 2 Current vs Source Current')
plt.savefig('column_2.png')
print("Maximum of column 2 current is: ", column_2.max(), "uA")
print("Maximum column 2 current occurs at: ", source[column_2.idxmax()], "A")
```

```python
#overlay both columns
plt.plot(source, column_1, label='Column 1')
plt.plot(source, column_2, label='Column 2')
plt.xlabel('Source Current (A)')
plt.ylabel('Column Current (uA)')
plt.title('Column Current vs Source Current')
plt.legend()
plt.savefig('columns.png')
```

```python
#overlay all 4
plt.plot(source, foil, label='Foil')
plt.plot(source, target, label='Target')
plt.plot(source, column_1, label='Column 1')
plt.plot(source, column_2, label='Column 2')
plt.xlabel('Source Current (A)')
plt.ylabel('Current (uA)')
plt.title('Current vs Source Current')
plt.legend()
plt.savefig('all.png')
```

```python
#calculating ratio of B field to source current for each data point
cyc_freq = 27.2e6
mp = 1.67e-27
e = 1.6e-19
max_current = source[column_2.idxmax()]

B = cyc_freq*2*3.1415*mp/e
print("Magnetic Field (B) is:" , B, "T")
```

```python
def field_to_current_ratio(f,m,e,I):
    return 2*f*math.pi*m/(e*I)
```

```python
ratio =  field_to_current_ratio(cyc_freq,mp,e,max_current)
print("Ratio of B field to source current is: ", ratio, "T/A")
```
